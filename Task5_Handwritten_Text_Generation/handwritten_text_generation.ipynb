```python
#CELL 1: IMPORT LIBRARIES
# CELL 1: IMPORT LIBRARIES
import random
```

```python
#CELL 2: LOAD TEXT DATA
# CELL 2: LOAD TEXT DATA
with open("input.txt", "r") as file:
    text = file.read().lower()

print(text[:300])
```

```python
#CELL 3: CHARACTER MAPPING
model = {}

for i in range(len(text) - 1):
    char = text[i]
    next_char = text[i + 1]
    
    if char not in model:
        model[char] = []
    model[char].append(next_char)
```

```python
# CELL 4: TEXT GENERATION FUNCTION
def generate_text(start_char, length=300):
    current = start_char
    result = current
    
    for _ in range(length):
        if current not in model:
            current = random.choice(list(model.keys()))
        else:
            current = random.choice(model[current])
        result += current
    
    return result
```

```python
# CELL 5: GENERATE HANDWRITTEN STYLE TEXT

print(generate_text("m"))
```

## Handwritten Text Generation

This project demonstrates handwritten-style text generation using a
character-level Markov Chain approach. The model learns character transition
patterns from textual data and generates new text sequences without using
deep learning frameworks.
